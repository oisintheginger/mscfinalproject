# Create for loop that calls request for each property:

```python
import time

import requests


#define function for taking api information

def retrieve_images2(zpid):
    url = "https://zillow-com1.p.rapidapi.com/images"

    headers = {
    "X-RapidAPI-Key": "0f64ef61f3msh8589bcbe6519721p1fd4adjsn7b6792b52973",
    "X-RapidAPI-Host": "zillow-com1.p.rapidapi.com"
    }
    querystring = {"zpid":zpid}
    response = requests.get(url, headers=headers, params=querystring)
    return response.json()

print(zpids[2])
```

```python
import time
import requests
import pandas as pd

rows_to_add = []
iterations = len(zpids)

for i in range(iterations):
    print("iteration" + str(i) + ", zpid: " + str(zpids[i]))
    temporary_dict = {}
    
    try:
        # Attempt to retrieve JSON data
        temporary_dict = retrieve_images2(zpids[i])
        
        if temporary_dict is not None and "images" in temporary_dict:
            # Check if the JSON data contains "images" key
            images = temporary_dict["images"]
            if images is not None:
                new_row = {}
                for image in range(len(images)):
                    url = images[image].replace("1536", "960")
                    new_row["zpid"] = zpids[i]
                    new_row["geoLocation"] = geoLocations[i]
                    new_row["image" + str(image)] = url
                rows_to_add.append(new_row)
        else:
            # Handle the case where the JSON response is not as expected
            error_message = "Invalid JSON response"
            new_row = {str(i): error_message}
            rows_to_add.append(new_row)
    except Exception as e:
        # Handle any exceptions that might occur during the request
        error_message = str(e)
        new_row = {str(i): error_message}
        rows_to_add.append(new_row)

    if i % 2 == 0 and i <= iterations - 1:
        # If we have made 2 requests and it's not the last iteration, sleep for 2 seconds
        time.sleep(2)

prop_image_df = pd.DataFrame(rows_to_add)

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Display the resulting DataFrame
```
