```python
from collections import Counter

# First list
first_list = ["Daft.ie", "Myhome.ie", "Immoweb.be", "Funda.nl", "Rent.ie", "Huurwoningen.nl", "Pararius.nl", "Universityliving.com", "Amberstudent.com", "Kamernet.nl"]

# Second list
second_list = ["Daft.ie", "rent.ie"]

# Flatten the first list to lowercase for case-insensitive comparison
flat_first_list = [word.lower() for sublist in map(str.split, first_list) for word in sublist]

# Count occurrences of words from the second list in the first list
word_counts = Counter(flat_first_list)

# Print the results
print("Occurrences of words from the second list in the first list:")
for word in second_list:
    count = word_counts[word.lower()]  # Convert to lowercase for case-insensitive comparison
    print(f"{word}: {count} times")
```

```python
from scipy.stats import spearmanr

# Sample data
data = {'Overall_Suitability': [4, 5, 3, 5, 2],
        'Price': [3, 4, 2, 5, 1],
        'Bed_Count': [2, 3, 1, 4, 2]}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate Spearman correlation coefficients and p-values
correlation_price, p_value_price = spearmanr(df['Overall_Suitability'], df['Price'])
correlation_bed_count, p_value_bed_count = spearmanr(df['Overall_Suitability'], df['Bed_Count'])

# Print the results
print(f"Spearman Correlation with Price: {correlation_price}")
print(f"P-value for Price: {p_value_price}")
print(f"Spearman Correlation with Bed Count: {correlation_bed_count}")
print(f"P-value for Bed Count: {p_value_bed_count}")
```

```python
import numpy as np
def spearman_confidence_interval(data1, data2, alpha=0.05):
    rho, p_value = spearmanr(data1, data2)
    z = np.arctanh(rho)
    sigma = 1 / np.sqrt(len(data1) - 3)
    margin_error = 1.96 * sigma
    ci_low, ci_high = np.tanh([z - margin_error, z + margin_error])
    return rho, p_value, ci_low, ci_high

correlation_price, p_value_price, ci_low_price, ci_high_price = spearman_confidence_interval(df['Overall_Suitability'], df['Price'])
correlation_bed_count, p_value_bed_count, ci_low_bed_count, ci_high_bed_count = spearman_confidence_interval(df['Overall_Suitability'], df['Bed_Count'])

print(f"Spearman Correlation with Price: {correlation_price} (CI: [{ci_low_price}, {ci_high_price}])")
print(f"P-value for Price: {p_value_price}")
print(f"Spearman Correlation with Bed Count: {correlation_bed_count} (CI: [{ci_low_bed_count}, {ci_high_bed_count}])")
print(f"P-value for Bed Count: {p_value_bed_count}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {'Overall_Suitability': [4, 5, 3, 5, 2],
        'Price': [3, 4, 2, 5, 1],
        'Bed_Count': [2, 3, 1, 4, 2]}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate the correlation matrix
correlation_matrix = df.corr(method='spearman')

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=.5)
plt.title('Spearman Correlation Matrix')
plt.show()
```
