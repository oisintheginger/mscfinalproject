```python
 import pandas as pd
```

```python
crime = pd.read_csv("D:/datasets/CRIME/unique_neighbours_crime.csv")
```

```python
unique_neighbourhoods = pd.read_csv("D:/datasets/DatabaseTables///NeighbourhoodsV4.csv")
```

```python
def convert_to_point(string_coordinates):
    # Split the string into latitude and longitude
    lat, lon = map(float, string_coordinates.split(', '))
    
    # Format the coordinates
    formatted_point = f"POINT({lat} {lon})"
    
    return formatted_point

crime['neighbourhoodGeoLocation'] = crime['neighbourhoodGeoLocation'].apply(convert_to_point)
```

```python
crime_coordinates = set(crime['neighbourhoodGeoLocation'])
neighbourhood_coordinates = set(unique_neighbourhoods['neighbourhoodGeoLocation'])

coordinates_not_in_neighbourhood = crime_coordinates - neighbourhood_coordinates
```

```python
filtered_crime = crime[crime['neighbourhoodGeoLocation'].isin(coordinates_not_in_neighbourhood)]
```
