Defining function to create the safe area tag based on the z score column

```python
#Defining a function to create the "safe area" tag based on the z score column
def create_safe_area_tag(row):
    # Get the z score from the row
    z_score = row['z_scores_before_mapping']
    # Check if the z score is less than or equal to 4
    if z_score <= 4:
        # Return "safe area" as the tag
        return 'safe area'
    else:
        # Return an empty string
        return ''
```

Applying the function to the dataframe and assigning the result to a new column

```python
crimescores['safe_area_tag'] = crimescores.apply(create_safe_area_tag, axis=1)
```

Defining function to create the services tag based on the z score column

```python
def create_nearby_service_tags(row):
    # Get the service columns from the row
    services = row[['bank_z_score_mapped', 'bar_z_score_mapped', 'beauty_salon_z_score_mapped', 'bus_station_z_score_mapped', 'cafe_z_score_mapped', 'fire_station_z_score_mapped', 'gym_z_score_mapped', 'hospital_z_score_mapped', 'night_club_z_score_mapped', 'park_z_score_mapped', 'pharmacy_z_score_mapped', 'police_station_z_score_mapped', 'restaurant_z_score_mapped', 'supermarket_z_score_mapped', 'train_station_z_score_mapped', 'transit_station_z_score_mapped']]
    # Sort the services by their z scores in descending order
    services = services.sort_values(ascending=False)
    # Get the top three service names as the tags
    tags = services.index[:3].tolist()
    # Return the tags
    return tags
```

Applying the function to the dataframe and assigning the result to a new column

```python
servicescore['nearby_service_tags'] = servicescore.apply(create_nearby_service_tags, axis=1)
```

generating threshold with sample dataset

```python
crimescores = pd.read_csv("crime_z_scores.csv")

# Calculating the threshold as the 90th percentile of z scores
threshold = np.percentile(crimescores['mapped_values_sigmoid'], 90)

# Defining a function to create the "safe area" tag based on the threshold
def create_safe_area_tag(row):
    # Getting the z score from the row
    z_score = row['mapped_values_sigmoid']
    # Checking if the z score is less than or equal to the calculated threshold
    if z_score >= threshold:
        # Returning "safe area" as the tag
        return 'safe area'
    else:
        # Returning an empty string
        return ''

# Applying the function to create the 'safe_area_tag' column
crimescores['safe_area_tag'] = crimescores.apply(create_safe_area_tag, axis=1)
```

service tags

```python
servicescore=pd.read_csv("service_scores_202311271059.csv")

def create_nearby_service_tags(row):
    # Get the service columns from the row
    services = row[['finance_score', 'transportation_score', 'personal_care_score', 'retail_score', 'fitness_score', 'leisure_score', 'emergency_score']]
    # Get the service with the highest z score
    highest_score_service = services.idxmax()
    # Return the service with the highest score as the tag
    return highest_score_service

# using the function on the DataFrame
tags_table['service_tag'] = servicescore.apply(create_nearby_service_tags, axis=1)
```

formatting the tags so it is more presentable to users when showed

```python
# Removing '_score' from all values in the 'safety_tag' column
tags_table['service_tag'] = tags_table['service_tag'].str.replace('_score', '')
```

```python
# Removing '_score' from all values in the 'safety_tag' column
tags_table['service_tag'] = tags_table['service_tag'].str.replace('emergency', 'emergency services')
```
