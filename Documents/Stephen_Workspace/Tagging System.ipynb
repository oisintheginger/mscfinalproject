Defining function to create the safe area tag based on the z score column

```python
#Defining a function to create the "safe area" tag based on the z score column
def create_safe_area_tag(row):
    # Get the z score from the row
    z_score = row['z_scores_before_mapping']
    # Check if the z score is less than or equal to 4
    if z_score <= 4:
        # Return "safe area" as the tag
        return 'safe area'
    else:
        # Return an empty string
        return ''
```

Applying the function to the dataframe and assigning the result to a new column

```python
crimescores['safe_area_tag'] = crimescores.apply(create_safe_area_tag, axis=1)
```

Defining function to create the services tag based on the z score column

```python
def create_nearby_service_tags(row):
    # Get the service columns from the row
    services = row[['bank_z_score_mapped', 'bar_z_score_mapped', 'beauty_salon_z_score_mapped', 'bus_station_z_score_mapped', 'cafe_z_score_mapped', 'fire_station_z_score_mapped', 'gym_z_score_mapped', 'hospital_z_score_mapped', 'night_club_z_score_mapped', 'park_z_score_mapped', 'pharmacy_z_score_mapped', 'police_station_z_score_mapped', 'restaurant_z_score_mapped', 'supermarket_z_score_mapped', 'train_station_z_score_mapped', 'transit_station_z_score_mapped']]
    # Sort the services by their z scores in descending order
    services = services.sort_values(ascending=False)
    # Get the top three service names as the tags
    tags = services.index[:3].tolist()
    # Return the tags
    return tags
```

Applying the function to the dataframe and assigning the result to a new column

```python
servicescore['nearby_service_tags'] = servicescore.apply(create_nearby_service_tags, axis=1)
```
