```python
import pandas as pd
```

```python
prop_details = pd.read_csv("D:/DataSets/backup_finished/property_details.csv")
```

```python
import pandas as pd

# Select columns from 7 to the end
selected_columns = prop_details.iloc[:, 6:-1]

my_dict = {}

for column in selected_columns.columns: 
    column_counts = selected_columns[column].value_counts()[1]
    print(column_counts)
    
    my_dict[column] = column_counts
```

```python
sorted_dict = dict(sorted(my_dict.items(), key=lambda item: item[1]))
sorted_dict
```

```python
import matplotlib.pyplot as plt


bar_names = list(sorted_dict.keys())
bar_counts = list(sorted_dict.values())

colors = ['red', 'green', 'blue', 'purple']
plt.barh(bar_names, bar_counts, color=colors, height=1, edgecolor='black', linewidth=1.5)  # Adjust height and width for longer and thicker bars

plt.xlabel('Bar Counts')
plt.ylabel('Bar Names')
plt.title('Sorted and Customized Horizontal Bar Chart from Dictionary')

for i, count in enumerate(bar_counts):
    plt.text(count + 0.2, i, str(count), ha='left', va='center', color='white', fontweight='bold')

plt.gcf().set_size_inches(10, 8)

# Show the plot
plt.show()
```
