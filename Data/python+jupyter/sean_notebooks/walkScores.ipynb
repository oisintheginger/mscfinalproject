```python
import time

import requests
    

iterations = len(school_zpids)
school_to_add = []
for i in range(iterations):
    
    print("iteration : " + str(i) + ", zpid = " + str(school_zpids["zpids"][i]))
    
    url = "https://zillow-com1.p.rapidapi.com/walkAndTransitScore"

    querystring = {"zpid":school_zpids["zpids"][i]}

    headers = {
        "X-RapidAPI-Key": "0f64ef61f3msh8589bcbe6519721p1fd4adjsn7b6792b52973",
        "X-RapidAPI-Host": "zillow-com1.p.rapidapi.com"
    }
    
    temp = {}
    temp = requests.get(url, headers=headers, params=querystring).json()

    if temp is not None:
        school_row = {}
        if temp.get("transitScore"):
            school_row["transit_score"] = temp["transitScore"]["transit_score"]
            school_row["transit_link"] = temp["transitScore"]["ws_link"]
            school_row["transit_description"] = temp["transitScore"]["description"]

        if temp.get("walkScore"):
            school_row["walk_score"] = temp["walkScore"]["walkscore"]
            school_row["walk_link"] = temp["walkScore"]["ws_link"]
            school_row["walk_description"] = temp["walkScore"]["description"]

        if temp.get("bikeScore"):
            school_row["bike_score"] = temp["bikeScore"]["bikescore"]
            school_row["bike_description"] = temp["bikeScore"]["description"]

        school_to_add.append(school_row)


        
    if i % 2 == 0 and i <= iterations:
        time.sleep(2)
        
nearby_school_df = pd.DataFrame(school_to_add)
```
